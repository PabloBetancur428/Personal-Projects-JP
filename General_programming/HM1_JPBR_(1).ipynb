#### HM1 for computer science MAIA applying always switch strategy.

- You are on a TV show and there are three doors in front of you:
 Two of them with a goat behind and one with a car;

- The host knows where the car is;

- I want you to simulate that you are a contestant on this show 1000 times.
 Each time, you pick a random door as your first choice,
 let the host open a door that reveals a goat,
 and then switch your choice to the door that the host didn't open.
With that strategy (known as the "always switch" strategy), how often do you win the car? Show me the percentage!

Here are a couple of details that I want to be clear:
- Before starting each game, the host randomly selects which door contains the car.

- After you make your initial choice, the host always opens one door that contains a goat,
and it will always be a door that you did not initially choose.
(If the host has two options for which door to open to reveal a goat, he will randomly select which of those two doors to open.)

- After the host opens a door that contains a goat,
you will always be given the option to switch your choice
to the door that is still closed, and you will always accept that option.

```python
#Import modules
import random

cars = 0
goats = 0
print("******************** WELCOME TO THE GOATCAR GAME! ********************")
print("Let's see how many times you get a car from this game applying always switch strategy! ")
for k in range(1000):
    print("**********WELCOME TO ITERATION {} OF GOATCAR GAME! **********".format(k+1))

    doors = [0,1,2]

    #Host will randomly select where is the car in each iteration
    host_choice = random.choice(range(0,3))
    doors[host_choice] = "car"

    # Since there is already a str in the list, we have to do a try-except statement
    #in order to avoid code-crash because an error and fill the other two doors with goats
    for i in range(len(doors)):
        try:
            if isinstance(doors[i],int):
                doors[i] = "goat"
        except:
            continue

    print("Let's try your luck!")
    print("Where is the car?????")
    print("""
        ???????????????????
         ___     ___     ___
        |   |   |   |   |   |
        | 1 | - | 2 |-  | 3 |
        |___|   |___|   |___|

        ????????????????????
    """)
    #User choice. This could be an input. Since there are 1000 iteration, it will be randomly selected.
    user_c = random.choice(range(0,3))

    #Get a list with a door (or doors in case you get the car in the first choice) corresponding to a goat. Besides that, It will avoid the selection from the user
    goat_indx = [i for i, x in enumerate(doors) if i != doors.index(doors[user_c]) and x != 'car']

    goat_door = random.choice(goat_indx)
    print("Now, I will show you a {} in door number... {}!!!".format(doors[goat_door],goat_door + 1))

    #Asks about changing the option, but the answer doesn't really matter since it will always be changed
    print("Would you like to change your option to the door that is still closed? [Y/N] ")
    print("I don't care, anyway I'll change it and you'd like that!")

    #Create the last switch, avoiding the first choice and the opened goat door
    last_switch = [i for i, x in enumerate(doors) if i != doors.index(doors[user_c]) and i != goat_door]

    #Just some basic counters to take the percentage at the end
    if doors[last_switch[0]] == "car":
        cars += 1
    else:
        goats += 1
    print("Now your door is {} !!!".format(last_switch[0] + 1))
    print("AND YOU WON A FANTASTIIIIIC.... {}!!!!!".format(doors[last_switch[0]].upper()))

print("Percentage of won cars: {}%".format(cars*100/1000))
print("Percentage of won goats: {}%".format(goats*100/1000))
print("******************** PROGRAM FINISHED ********************")
```
